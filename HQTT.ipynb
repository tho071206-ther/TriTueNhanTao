```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# --- [BÀI 5] ỨNG DỤNG HỒI QUY TUYẾN TÍNH ---
def app_house_price_prediction():
    print("--- APP: DỰ BÁO GIÁ NHÀ ĐẤT ---")

    # Dữ liệu mẫu: Diện tích (m2) -> Giá (Tỷ VND)
    X_area = np.array([[30], [40], [50], [60], [70], [80], [100], [120]])
    y_price = np.array([1.5, 2.0, 2.8, 3.0, 3.5, 4.2, 5.0, 6.1])

    # Huấn luyện
    model = LinearRegression()
    model.fit(X_area, y_price)

    # Demo dự báo
    new_house = [[90], [150]]
    predicted = model.predict(new_house)

    for area, price in zip(new_house, predicted):
        print(f"Nhà rộng {area[0]}m2 dự kiến có giá: {price:.2f} tỷ VND")

    # Trực quan hóa
    plt.scatter(X_area, y_price, color='blue', label='Dữ liệu thực tế')
    plt.plot(X_area, model.predict(X_area), color='red', label='Đường xu hướng')
    plt.scatter(new_house, predicted, color='green', s=100, marker='*', label='Dự báo mới')
    plt.xlabel('Diện tích (m2)')
    plt.ylabel('Giá (Tỷ VND)')
    plt.title('Mô hình định giá bất động sản')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    app_house_price_prediction()
```
