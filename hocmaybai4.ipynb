```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.neighbors import NearestNeighbors

def app_customer_segmentation():
    print("\n--- APP 1: PHÂN KHÚC KHÁCH HÀNG (K-MEANS) ---")
    # Dữ liệu: [Thu nhập (k$), Điểm chi tiêu (1-100)]
    X = np.array([[15, 39], [15, 81], [16, 6], [16, 77], [17, 40],
                  [30, 20], [35, 25], [40, 60], [45, 65], [50, 70],
                  [70, 90], [80, 95], [90, 85], [85, 80], [75, 85]])

    kmeans = KMeans(n_clusters=3, n_init='auto').fit(X)

    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap='rainbow')
    plt.title('Phân nhóm khách hàng')
    plt.xlabel('Thu nhập'); plt.ylabel('Điểm chi tiêu')
    print("Đã phân cụm khách hàng xong (Xem biểu đồ).")

def app_product_recommender():
    print("\n--- APP 2: GỢI Ý ĐIỆN THOẠI (K-NN) ---")
    # Data: [RAM, Pin, Giá]
    phones = np.array([[4, 3000, 3], [8, 4500, 7], [12, 5000, 15],
                       [8, 4400, 7.5], [4, 3100, 3.2]])
    names = ['Phone A', 'Phone B', 'Phone C', 'Phone D', 'Phone E']

    # Tìm 2 sản phẩm gần nhất (bao gồm chính nó)
    nbrs = NearestNeighbors(n_neighbors=2).fit(phones)
    dist, ind = nbrs.kneighbors([phones[1]]) # Tìm giống Phone B

    print(f"Sản phẩm đang xem: {names[1]}")
    print(f"Sản phẩm gợi ý: {names[ind[0][1]]} (Do cấu hình tương tự)")

    plt.subplot(1, 2, 2)
    plt.bar(names, phones[:, 2], color=['gray', 'blue', 'gray', 'cyan', 'gray'])
    plt.title('So sánh giá các sản phẩm')

if __name__ == "__main__":
    app_customer_segmentation()
    app_product_recommender()
    plt.show()
```
